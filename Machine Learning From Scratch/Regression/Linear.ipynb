```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#generate data
np.random.seed(123)
N=20
beta0=-4
beta1=2
x=np.random.randn(N)
e=np.random.randn(N)
y=beta0+beta1*x+e
print('x= ',x)
print('y= ',y)
```

```python
true_x=np.linspace(min(x),max(x),100)
print(true_x)
```

```python
true_y=beta0+beta1*true_x
print(true_y)
```

```python
#ploting the graph
fig, ax = plt.subplots()
sns.scatterplot(x, y, s = 40, label = 'Data')
sns.lineplot(true_x,true_y,color='red',label='True Model')
ax.set_title(fr"$y = {beta0} + ${beta1}$x + \epsilon$", fontsize = 16)
ax.set_xlabel('x', fontsize = 14)
ax.set_ylabel('y', fontsize=14, rotation=0, labelpad=10)
ax.legend(loc = 4)
sns.despine()
```

```python
#estimating the model
beta1_hat=sum((x-np.mean(x))*(y-np.mean(y)))/sum((x-np.mean(x))**2)
beta0_hat=np.mean(y)-beta1_hat*np.mean(x)
fit_y=beta0_hat+beta1_hat*true_x
fig, ax=plt.subplots()
sns.scatterplot(x,y,s=40,label='Data')
sns.lineplot(true_x,true_y,color='red',label='True Model')
sns.lineplot(true_x,fit_y,color='blue',label='Estimated Model')
ax.set_title(fr"$y={beta0}+${beta1}$x+ \epsilon$",fontsize=16)
ax.set_xlabel('x', fontsize = 14)
ax.set_ylabel('y', fontsize=14, rotation=0, labelpad=10)
ax.legend(loc = 4)
sns.despine()
```

contructing the linear regression with numpy only

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

class LinearRegression:
    # record data and dimensions
    def fit(self,X,y,intercept=False):
        if intercept==False: # add intercept (if not already included)
            ones=np.ones(len(X)).reshape(len(X),1) #column of ones
            X=np.concatenate((ones,X),axis=1)
        self.X=np.array(X)
        self.y=np.array(y)
        self.N, self.D=self.X.shape
        
        #estimate parameters
        self.beta_hats=np.dot(np.linalg.inv(np.dot(self.X.T,self.X)),np.dot(self.X.T,self.y))
        
        #make-in samples prediction
        self.y_hat=np.dot(self.X,self.beta_hats)
        
        #calculate loss
        self.L=1/2*(np.sum((self.y-self.y_hat)**2))
        
    def predict(self,X_test, intercept=True):
        
        #from predictions
        self.y_test_hat=np.dot(X_test,self.beta_hats)
            
```

```python
#importing the datasets
from sklearn import datasets
boston=datasets.load_boston()
X=boston['data']
y=boston['target']
```

```python
#fitting the model
lr=LinearRegression()
lr.fit(X,y,intercept=False)
```

```python
fig, ax = plt.subplots()
sns.scatterplot(lr.y, lr.y_hat)
ax.set_xlabel(r'$y$', size = 16)
ax.set_ylabel(r'$\hat{y}$', rotation = 0, size = 16, labelpad = 15)
ax.set_title(r'$y$ vs. $\hat{y}$', size = 20, pad = 10)
sns.despine()
```

Implementation of linear regression with SKLEARN

```python
from sklearn import datasets
import seaborn as sns
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
boston=datasets.load_boston()
X_train=boston['data']
y_train=boston['target']

model=LinearRegression()
model.fit(X_train,y_train);
```

```python
prediction=model.predict(X_train)
```

```python
fig,ax=plt.subplots()

sns.scatterplot(y_train,prediction)
ax.set_xlabel(r'$y$', size = 16)
ax.set_ylabel(r'$\hat{y}$', rotation = 0, size = 16, labelpad = 15)
ax.set_title(r'$y$ vs. $\hat{y}$', size = 20, pad = 10)
sns.despine()
```

```python
predictors=boston.feature_names
beta_hats=model.coef_
print('\n'.join([f'{predictors[i]}: {round(beta_hats[i], 3)}' for i in range(3)]))
```

Linear Regression using Statsmodel

```python
import pandas as pd
df=pd.DataFrame(X_train,columns=boston['feature_names'])
df['target']=y_train
display(df.head())
```

```python
formula = 'target ~ ' + ' + '.join(boston['feature_names'])
print('formula:', formula)
```

```python
import statsmodels.formula.api as smf

sm_model=smf.ols(formula,data=df)
sm_fit=sm_model.fit()
sm_pred=sm_fit.predict(df)
```

```python
fig,ax=plt.subplots()

sns.scatterplot(df['target'],sm_pred)
ax.set_xlabel(r'$y$', size = 16)
ax.set_ylabel(r'$\hat{y}$', rotation = 0, size = 16, labelpad = 15)
ax.set_title(r'$y$ vs. $\hat{y}$', size = 20, pad = 10)
sns.despine()
```
