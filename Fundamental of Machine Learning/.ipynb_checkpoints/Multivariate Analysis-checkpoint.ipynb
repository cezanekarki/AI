```python
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#loading dataset
iris = datasets.load_iris()

#converting the data into dataframe
iris = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                   columns=iris['feature_names']+['species'])
iris.species = np.where(iris.species == 0.0, 'setosa', np.where(iris.
species==1.0,'versicolor', 'virginica'))

iris.columns = iris.columns.str.replace(' ','')
iris.describe()
```

```python
#print the mean of each column by species
iris.groupby(by='species').mean()
```

```python
#plot the mean of each feature for each label class
iris.groupby(by='species').mean().plot(kind="bar")
plt.title("Class VS Measurement")
plt.ylabel("Mean measurement")
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
```

```python
#correlation matrix

import statsmodels.api as sm
corr=iris.corr()
print(corr)

sm.graphics.plot_corr(corr, xnames=list(corr.columns))
```

```python
#pair plot
from pandas.plotting import scatter_matrix

scatter_matrix(iris,figsize=(10,10))
plt.suptitle("Pair Plot")
```

#Findings from eda
1) There are no missing values.
2) Sepal is Longer than Petal.
3) Sepal is also wider than Petal
4) Average petal length of setosa is much smaller than versicolor and
    virginica; however the average sepal width of setosa is higher than
    versicolor and virginica.
5) Petal length and width are strongly correlated, that is, 96% of the
    time width increases with increase in length.
6) Petal length has negative correlation with sepal width, that is, 42%
    of the time increase in sepal width will decrease petal length.
7) Initial conclusion from data: Based on length and width of sepal/
    petal alone, you can conclude that versicolor/virginica might
    resemble in size; however setosa characteristics seem to be
    noticeably different from the other two.
